## Purpose Encodings

```python
class Purpose:
    def __init__(self, name, id, parent_id, code, aip_code = 0, pip_code = 0):
        self.name = name
        self.id = id
        self.parent_id = parent_id
        self.code = code
        self.aip_code = aip_code
        self.pip_code = pip_code

    def __repr__(self):
        return f"Name: {self.name}, ID: {self.id}, Parent ID: {self.parent_id}, Code: {self.code}, AIP Code: {self.aip_code}, PIP Code: {self.pip_code}"

    def to_dict(self):
        return {
            "name": self.name,
            "id": self.id,
            "parent_id": self.parent_id,
            "code": self.code,
            "aip_code": self.aip_code,
            "pip_code": self.pip_code
        }
```

```python
def decode_purposes(encoded_value, purposes):
    decoded_purposes = []
    for purpose in purposes.values():
        if encoded_value & purpose.code:
            decoded_purposes.append(purpose.name)
    return decoded_purposes

test = 6244

decode_purposes(test, purposes)
```

```python
import random

def logical_or(codes):
    result = 0
    for code in codes:
        result = result | code
    return result

def pick_k_research_third_party_aip(k = 3):
    return random.choices(
        [purposes.get('public_research').aip_code, 
         purposes.get('private_research').aip_code, 
         purposes.get('military_research').aip_code, 
         purposes.get('non_military_research').aip_code, 
         purposes.get('third_party').aip_code, 
         purposes.get('marketing').aip_code, 
         purposes.get('product_development').aip_code],k = k)

def pick_k_research_third_party_pip(k = 2):
    return random.choices(
        [purposes.get('public_research').pip_code, 
         purposes.get('private_research').pip_code, 
         purposes.get('military_research').pip_code, 
         purposes.get('non_military_research').pip_code,
         purposes.get('third_party').pip_code, 
         purposes.get('marketing').pip_code, 
         purposes.get('product_development').pip_code], k = k)


# AIP Assignment Functions

def assign_report_year_aip():
    return purposes.get('general_purpose').aip_code

def assign_diagnosis_treatment_determination_type_aip():
    aip = [purposes.get('clinical_care').aip_code, purposes.get('communication').aip_code]
    aip.extend(pick_k_research_third_party_aip(3))

    return logical_or(aip)

def assign_age_range_gender_aip():
    aip = [purposes.get('clinical_care').aip_code, purposes.get('patient_support_service').aip_code]
    aip.extend(pick_k_research_third_party_aip(4))
    return logical_or(aip)

def assign_findings_aip():
    aip = [purposes.get('clinical_care').aip_code, purposes.get('communication').aip_code]
    aip.extend(pick_k_research_third_party_aip(1))
    return logical_or(aip)

def assign_name_phone_address_aip():
    aip = [purposes.get('clinical_care').aip_code, purposes.get('patient_support_service').aip_code]
    aip.extend(pick_k_research_third_party_aip(3))
    return logical_or(aip)

def assign_age_blood_type_aip():
    aip = [purposes.get('clinical_care').aip_code]
    aip.extend(pick_k_research_third_party_aip(3))
    return logical_or(aip)

def assign_insurance_aip():
    aip = [purposes.get('patient_support_service').aip_code]
    return logical_or(aip)

def assign_consulting_physician_aip():
    aip = [purposes.get('clinical_care').aip_code, purposes.get('patient_support_service').aip_code]
    return logical_or(aip)


# PIP Assignment Functions
def assign_report_year_pip():
    return 0

def assign_diagnosis_treatment_determination_type_pip():
    pip = [purposes.get('marketing').pip_code]
    pip.extend(pick_k_research_third_party_pip(4))

    return logical_or(pip)

def assign_age_range_gender_pip():
    pip = []
    pip.extend(pick_k_research_third_party_pip(1))
    return logical_or(pip)

def assign_findings_pip():
    pip = [purposes.get('marketing').pip_code]
    pip.extend(pick_k_research_third_party_pip(3))
    return logical_or(pip)

def assign_name_phone_address_pip():
    pip = []
    pip.extend(pick_k_research_third_party_pip(1))
    return logical_or(pip)

def assign_age_blood_type_pip():
    pip = []
    pip.extend(pick_k_research_third_party_pip(3))
    return logical_or(pip)

def assign_insurance_pip():
    pip = [purposes.get('third_party').pip_code]
    return logical_or(pip)

def assign_consulting_physician_pip():
    pip = []
    return logical_or(pip)
```

```python
# Populate database
for i in range(len(df)):
    doc = df.iloc[i].to_dict()

    diagnosis_treatment_determination_type_aip = assign_diagnosis_treatment_determination_type_aip()
    name_phone_address_aip = assign_name_phone_address_aip()
    age_range_gender_aip = assign_age_range_gender_aip()
    age_blood_type_aip = assign_age_blood_type_aip()
    insurance_aip = assign_insurance_aip()
    doc['AIP'] = {'ReportYear': assign_report_year_aip(),
                    'DiagnosisCategory': diagnosis_treatment_determination_type_aip,
                    'DiagnosisSubCategory': diagnosis_treatment_determination_type_aip,
                    'TreatmentCategory': diagnosis_treatment_determination_type_aip,
                    'TreatmentSubCategory': diagnosis_treatment_determination_type_aip,
                    'Determination': diagnosis_treatment_determination_type_aip,
                    'Type': diagnosis_treatment_determination_type_aip,
                    'Findings': assign_findings_aip(),
                    'AgeRange': age_range_gender_aip,
                    'PatientAge': age_blood_type_aip,
                    'PatientGender': age_range_gender_aip,
                    'PatientName': name_phone_address_aip,
                    'PatientPhone': name_phone_address_aip,
                    'PatientAddress': name_phone_address_aip,
                    'PatientBloodType': age_blood_type_aip,
                    'PatientSSN': insurance_aip,
                    'PatientInsuranceProvider': insurance_aip,
                    'PatientInsuranceNumber': insurance_aip,
                    'ConsultingPhysician': assign_consulting_physician_aip()    
                    }

    diagnosis_treatment_determination_type_pip = assign_diagnosis_treatment_determination_type_pip()
    name_phone_address_pip = assign_name_phone_address_pip()
    age_range_gender_pip = assign_age_range_gender_pip()
    age_blood_type_pip = assign_age_blood_type_pip()
    insurance_pip = assign_insurance_pip()

    doc['PIP'] = {
        'ReportYear': assign_report_year_pip(),
        'DiagnosisCategory': diagnosis_treatment_determination_type_pip,
        'DiagnosisSubCategory': diagnosis_treatment_determination_type_pip,
        'TreatmentCategory': diagnosis_treatment_determination_type_pip,
        'TreatmentSubCategory': diagnosis_treatment_determination_type_pip,
        'Determination': diagnosis_treatment_determination_type_pip,
        'Type': diagnosis_treatment_determination_type_pip,
        'Findings': assign_findings_pip(),
        'AgeRange': age_range_gender_pip,
        'PatientAge': age_blood_type_pip,
        'PatientGender': age_range_gender_pip,
        'PatientName': name_phone_address_pip,
        'PatientPhone': name_phone_address_pip,
        'PatientAddress': name_phone_address_pip,
        'PatientBloodType': age_blood_type_pip,
        'PatientSSN': insurance_pip,
        'PatientInsuranceProvider': insurance_pip,
        'PatientInsuranceNumber': insurance_pip,
        'ConsultingPhysician': assign_consulting_physician_pip()
    }
    collection.insert_one(doc)
```
