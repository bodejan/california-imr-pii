## Assign Purposes to Values in PII Dataset

```python
import pandas as pd

df = pd.read_csv('data/pii.csv')
df
```

```python
import random

purpose_hierachy = {
    "general_purpose": [
        "General-Purpose", "Clinical-Care", "Research", "Public-Research", "Military-Research",
        "Non-Military-Research", "Private-Research", "Patient-Support-Service", "Billing",
        "Communication", "Third-Party", "Marketing", "Product-Development"
    ],
    "clinical_care": ["Clinical-Care"],
    "research": [
        "Research", "Public-Research", "Military-Research", "Non-Military-Research", "Private-Research"
    ],
    "public_research": ["Public-Research", "Military-Research", "Non-Military-Research"],
    "private_research": ["Private-Research"],
    "military_research": ["Military-Research"],
    "non_military_research": ["Non-Military-Research"],
    "patient_support_service": ["Patient-Support-Service", "Billing", "Communication"],
    "billing": ["Billing"],
    "communication": ["Communication"],
    "third_party": ["Third-Party", "Marketing", "Product-Development"],
    "marketing": ["Marketing"],
    "product_development": ["Product-Development"]
}

def flatten_set(lst: list):
    return list(set([item for sublist in lst for item in sublist]))

def remove_intersection(list1, list2):
    """Removes elements that are present in both lists. Returns list1 with elements removed."""
    # Filter list1 to remove elements that are present in list2
    result = [item for item in list1 if item not in list2]
    
    return result



def pick_k_research_third_party(k=3):
    # Randomly select k purposes from the list of possible purposes
    selected_purposes = random.choices(
        [
            purpose_hierachy.get('research'),
            purpose_hierachy.get('public_research'), 
            purpose_hierachy.get('private_research'), 
            purpose_hierachy.get('military_research'), 
            purpose_hierachy.get('non_military_research'), 
            purpose_hierachy.get('third_party'), 
            purpose_hierachy.get('marketing'), 
            purpose_hierachy.get('product_development')
        ], k=k
    )
    
    return flatten_set(selected_purposes)

pick_k_research_third_party()
```

```python
def assign_report_year():
    selected_purposes = purpose_hierachy.get('general_purpose')
    prohibited_purpose = []

    return remove_intersection(selected_purposes, prohibited_purpose)

def assign_diagnosis_treatment_determination_type():
    selected_purposes = flatten_set([purpose_hierachy.get('clinical_care'), purpose_hierachy.get('communication')])
    selected_purposes.extend(pick_k_research_third_party())
    prohibited_purpose = flatten_set([purpose_hierachy.get('marketing')])

    return remove_intersection(selected_purposes, prohibited_purpose)

def assign_age_range_gender():
    selected_purposes = flatten_set([purpose_hierachy.get('clinical_care'), purpose_hierachy.get('patient_support_service')])
    selected_purposes.extend(pick_k_research_third_party())
    prohibited_purpose = []

    return remove_intersection(selected_purposes, prohibited_purpose)

def assign_findings():
    selected_purposes = flatten_set([purpose_hierachy.get('clinical_care'), purpose_hierachy.get('communication')])
    selected_purposes.extend(pick_k_research_third_party())
    prohibited_purpose = flatten_set([purpose_hierachy.get('marketing')])

    return remove_intersection(selected_purposes, prohibited_purpose)

def assign_name_phone_address():
    selected_purposes = flatten_set([purpose_hierachy.get('clinical_care'), purpose_hierachy.get('patient_support_service')])
    selected_purposes.extend(pick_k_research_third_party())
    prohibited_purpose = []

    return remove_intersection(selected_purposes, prohibited_purpose)

def assign_age_blood_type():
    selected_purposes = flatten_set([purpose_hierachy.get('clinical_care')])
    selected_purposes.extend(pick_k_research_third_party())
    prohibited_purpose = []

    return remove_intersection(selected_purposes, prohibited_purpose)

def assign_insurance():
    selected_purposes = flatten_set([purpose_hierachy.get('patient_support_service')])
    prohibited_purpose = []

    return remove_intersection(selected_purposes, prohibited_purpose)

def assign_consulting_physician():
    selected_purposes = flatten_set([purpose_hierachy.get('clinical_care'), purpose_hierachy.get('patient_support_service')])
    prohibited_purpose = []

    return remove_intersection(selected_purposes, prohibited_purpose)
```

## Populate MongoDB

```python
from pymongo.mongo_client import MongoClient
from pymongo.server_api import ServerApi
from dotenv import load_dotenv
import os

load_dotenv()

# Create a new client and connect to the server
client = MongoClient(os.getenv("DB_URI"), server_api=ServerApi('1'))

# Send a ping to confirm a successful connection
try:
    client.admin.command('ping')
    print("Pinged your deployment. You successfully connected to MongoDB!")
except Exception as e:
    print(e)
```

```python
db = client.get_database(os.getenv("DB_NAME"))
collection = db.get_collection(os.getenv("DB_COLLECTION2"))
```

```python
# Populate database
for i in range(len(df)):
    doc = df.iloc[i].to_dict()

    # Assign purposes using the new functions
    diagnosis_treatment_determination_type = assign_diagnosis_treatment_determination_type()
    name_phone_address = assign_name_phone_address()
    age_range_gender = assign_age_range_gender()
    age_blood_type = assign_age_blood_type()
    insurance = assign_insurance()

    # Store the purposes directly in the document with the suffix _IP (uppercase)
    doc['ReportYear_IP'] = assign_report_year()
    doc['DiagnosisCategory_IP'] = diagnosis_treatment_determination_type
    doc['DiagnosisSubCategory_IP'] = diagnosis_treatment_determination_type
    doc['TreatmentCategory_IP'] = diagnosis_treatment_determination_type
    doc['TreatmentSubCategory_IP'] = diagnosis_treatment_determination_type
    doc['Determination_IP'] = diagnosis_treatment_determination_type
    doc['Type_IP'] = diagnosis_treatment_determination_type
    doc['Findings_IP'] = assign_findings()
    doc['AgeRange_IP'] = age_range_gender
    doc['PatientAge_IP'] = age_blood_type
    doc['PatientGender_IP'] = age_range_gender
    doc['PatientName_IP'] = name_phone_address
    doc['PatientPhone_IP'] = name_phone_address
    doc['PatientAddress_IP'] = name_phone_address
    doc['PatientBloodType_IP'] = age_blood_type
    doc['PatientSSN_IP'] = insurance
    doc['PatientInsuranceProvider_IP'] = insurance
    doc['PatientInsuranceNumber_IP'] = insurance
    doc['ConsultingPhysician_IP'] = assign_consulting_physician()

    # Insert the document into the collection
    collection.insert_one(doc)
```
